```python
# create video from predicted frames
import cv2
import numpy as np
import glob
 
img_array = []
for filename in glob.glob('/content/MDnet/results/crossing/*.jpg'):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
    
    
 
 
out = cv2.VideoWriter('project_last.avi',cv2.VideoWriter_fourcc(*'DIVX'), 5, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
